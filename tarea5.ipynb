1. Sea $f(x) = −𝑥^3 − cos(x)$ y $p_0 = −1$. Use el método de Newton y de la Secante para encontrar $p_{2}$. ¿Se podría usar $p_{0} = 0$?

    El método de newton utiliza la siguiente fórmula iterativa \
                $p_n​ = p_{n−1} − \frac{f(p_{n−1}​)}{f'(p_{n−1})​}$

    Su derivada es: $f'(x) = -3x^2 + sin(x)$

    Con el valor $p_0 = 0$ \
        $f'(0) = -3(0)^2 + sin(0) = 0 + 0 = 0$

```python
import numpy as np
from scipy import optimize

# función y su derivada
def f(x):
    print(f" -> Iteración evaluando en p = {x:.6f}")
    return -x**3 - np.cos(x)

def df(x):
    return -3*x**2 + np.sin(x)


p0 = -1.0

print(" Método de Newton ")
p2_newton, info_newton = optimize.newton(f, x0=p0, fprime=df, maxiter=2,full_output=True, disp=False)
print(f" El valor de p2: {p2_newton:.6f}")


print("\n Método de la Secante ")
# La secante estricta necesita x0 y x1. 
# calcularemos x1 con un paso de Newton.
p1 = p0 -f(p0) / df(p0)
p2_secante, info_secante = optimize.newton(f, x0=p0, x1=p1,maxiter=1,full_output=True, disp=False)
print(f" El valor de p2: {p2_secante:.6f}")
```

5. La función $f(x) = tan(\pi x)  - 6$ tiene cero en $\frac{1}{\pi}$ arcontangente $6 \approx 0.447431543$. Sea $p_0 = 0$ y $p_1 = 0.48$ y use 10 iteraciones en cada uno de los siguientes métodos para aproximar esta raíz. ¿Cuál método es más eficaz y porqué? \
    a. Método de bisección \
    b. Método de Newton \
    c. Método de la secante

```python
def f_5(x):
    return np.tan(np.pi * x) - 6 

def df_5(x):
    return np.pi / (np.cos(np.pi * x)**2) 

# raiz exacta
raiz_exacta = np.arctan(6) / np.pi
print(f" Raíz teórica exacta: {raiz_exacta:.9f}\n")

print("\n  --- Método de la bisección ")
a = 0.0 
b = 0.48
for i in range(1,11):
    p = (a + b)/ 2.0
    error = abs(p - raiz_exacta)
    print(f" Iteración {i:2d}: p = {p:.9f} | Error: {error:.9f}")
    if f_5(a) * f_5(p) < 0:
        b = p
    else:
        a = p


print("\n  --- Método de Newton ")
p_newton = 0.0
for i in range(1,11):
    p_newton = p_newton - f_5(p_newton) / df_5(p_newton)
    error = abs(p_newton - raiz_exacta)
    print(f" Iteracion {i:2d}: p = {p_newton:-9f} | Error: {error:.9f}")


print("\n  --- Método de la Secante ")
p0_sec = 0.0
p1_sec = 0.48
for i in range(1,11):
    p_sig = p1_sec - f_5(p1_sec) * (p1_sec - p0_sec) / (f_5(p1_sec) - f_5(p0_sec))
    error = abs(p_sig - raiz_exacta)
    print(f" Iteracion {i:2d}: p = {p_sig:-9f} | Error: {error:.9f}")
    p0_sec = p1_sec
    p1_sec = p_sig
```

El método más eficaz es de la Bisección porque Newton y Secante divergen con esos puntos iniciales. Bisección garantiza convergencia siempre que f cambie de signo en el intervalo.

6. La función descrita por $f(x) = ln(x^2 +1) - e^{0.4x} cos(\pi  x)$ tiene un número infinito de ceros. \
    a. Determine, dentro de $10^{-6}$, el único cero negativo. \
    b. Determine, dentro de $10^{-6}$, los cuatro ceros positivos más pequeños. \
    c. Determine una aproximación inicial razonable para encontrar el enésimo cero positivo más pequeño de f.  [_Sugerencia:_ Dibuje una gráfica aproximada de f]\
    d. Use la parte c) para determinar, dentro de $10^{-6}$, el vigesimoquinto cero positivo más pequeño de f.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize


def f_6(x):
    return np.log(x**2 + 1) - np.exp(0.4 * x) * np.cos(np.pi * x)


def df_6(x):
    term1 = (2 * x) / (x**2 + 1)
    term2 = -0.4 * np.exp(0.4 * x) * np.cos(np.pi * x)
    term3 = np.pi * np.exp(0.4 * x) * np.sin(np.pi * x)
    return term1 + term2 + term3



# Generar valores de x para graficar desde -2 hasta 5
x_vals = np.linspace(-2, 10, 400)
y_vals = f_6(x_vals) 

# Dibujamos la gráfica principal
plt.figure(figsize=(14, 7)) 
plt.plot(x_vals, y_vals, label="f(x)", color="#1f77b4", linewidth=2)
plt.axhline(0, color='black', linewidth=1.5) # Eje X

# para mostrar en intervalos de 0.5
plt.xticks(np.arange(-2, 10.5, 0.5))


plt.grid(True, linestyle='--', alpha=0.8)
plt.title("Gráfica de f(x) = ln(x^2 + 1) - e^{0.4x} cos(\pi x)", fontsize=14, fontweight='bold')
plt.xlabel("Eje X ", fontsize=12)
plt.ylabel("f(x)", fontsize=12)
plt.legend()


plt.show()
```

```python
import numpy as np
from scipy import optimize

print("--- Literal a) ")
po_negativo = -0.4 # estimación inicial
raiz_negativo = optimize.newton(f_6, x0=po_negativo, fprime=df_6, tol=1e-6)
print(f" Raíz negativa: {raiz_negativo:.6f}")


print("\n--- Literal b) ")
estimaciones = [0.5, 1.5, 2.5, 3.5] # para los siguientes 4 puntos
for i, p0 in enumerate(estimaciones, 1):
    raiz = optimize.newton(f_6, x0=p0,fprime=df_6, tol=1e-6)
    print(f" Cero positivo {i}: {raiz:.6f}")


print("\n--- Literal c) ")
po_positivo = 8.5 # estimación inicial
raiz = optimize.newton(f_6, x0=po_positivo, fprime=df_6, tol=1e-6)
print(f" Raíz negativa: {raiz:.6f}")

```

Para encontrar el enésimo cero positivo, igualamos la función a cero:

$$\ln(x^2 + 1) = e^{0.4x} \cos(\pi x)$$

A medida que $x$ crece, el término $e^{0.4x}$ crece exponencialmente mucho más rápido que el logaritmo. Por lo tanto, para que la igualdad se mantenga, el término $\cos(\pi x)$ debe tender a cero.

Sabemos que $\cos(\theta)=0$ cuando $\theta$ es un múltiplo impar de $\pi/2$. 

Al despejar:

$$\pi x = \frac{(2n - 1)\pi}{2}$$

Obtenemos la fórmula general para la aproximación inicial del enésimo cero:

$$p_0 = n - 0.5$$


```python
print("\n--- Literal d) ")
n = 25 
p0_d = n - 0.5
raiz_d = optimize.newton(f_6, x0=p0_d, fprime=df_6, tol=1e-6)
print(f" Raíz negativa: {raiz_d:.6f}")
```

7. La función $f(x) = x^{\frac{1}{3}}$ tiene la raíz en $x=0$. Usando el punto de inicio de $x=1$ y $p_0=5$, $p_1 = 0.5$ para el método de secante, compare los resultados de los métodos de la secante y de Newton.

```python
import numpy as np
def f(x):
    return np.cbrt(x)

def df(x):
    if x == 0:
        return float('inf')
    return 1.0 / (3.0 * np.cbrt(x)**2)


p_newton = 1.0
print(" --- Método de Newton")
for i in range(1, 11):
    p_newton = p_newton - f(p_newton) / df(p_newton)
    print(f"Iteración {i:2d} p = {p_newton:12.6f}")

print("\n --- Método de la Secante")
p0_sec = 5.0
p1_sec = 0.5
for i in range(1, 11):
    denominador = f(p1_sec) - f(p0_sec)
    
    p_siguiente = p1_sec - f(p1_sec) * (p1_sec - p0_sec) / denominador
    print(f"Iteración {i:2d} p = {p_siguiente:12.6f}")
    

    p0_sec = p1_sec
    p1_sec = p_siguiente
```

El método de Newton no puede converger cuando x = 0, porque tiende a infinito.
El método de la Secante diverge, pero oscila entre los valores $\pm 3.47$ y $\pm 0.81$
Ninguno de los dos métodos sirve para raíces donde la derivada no existe o se anula.
